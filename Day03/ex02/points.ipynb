```python

import pandas as pd
import matplotlib.pyplot as plt


df_train = pd.read_csv('../resources/Train_knight.csv')
df_test = pd.read_csv('../resources/Test_knight.csv')

features: list[tuple] = [
            ("Empowered", "Prescience"),
            ("Deflection", "Survival"),
        ]

grouped = df_train.groupby("knight")
jedi = grouped.get_group("Jedi")
sith = grouped.get_group("Sith")
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))


for i, feature in enumerate(features):
    jediX, jediY = jedi[feature[0]], jedi[feature[1]]
    sithX, sithY = sith[feature[0]], sith[feature[1]]
    knightX, knightY = df_test[feature[0]], df_test[feature[1]]

    axes[0, i].scatter(
        jediX, jediY, color="blue", s=30, alpha=0.4, edgecolors="white"
    )
    axes[0, i].scatter(
        sithX, sithY, color="red", s=30, alpha=0.4, edgecolors="white"
    )
    axes[0, i].set_xlabel(feature[0])
    axes[0, i].set_ylabel(feature[1])
    axes[0, i].legend(labels=("Jedi", "Sith"))

    axes[1, i].scatter(
        knightX, knightY, color="green", s=30, alpha=0.4, edgecolors="white"
    )
    axes[1, i].set_xlabel(feature[0])
    axes[1, i].set_ylabel(feature[1])
    axes[1, i].legend(labels=("Knight",))

plt.tight_layout()
plt.show()
```
