```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('../resources/Train_knight.csv')
# data = pd.read_csv('../resources/Test_knight.csv')

print(data.head(2))
for col in data.columns:
    if not pd.api.types.is_string_dtype(data[col]):
        data[col] = (data[col] - data[col].min()) / (
                    data[col].max() - data[col].min()
                )
print(data.head(2))

if 'knight' in data:
    grouped = data.groupby("knight")
    jedi_data = grouped.get_group("Jedi")
    sith_data = grouped.get_group("Sith")
    js, jp = jedi_data["Deflection"], jedi_data["Survival"]
    ss, sp = sith_data["Deflection"], sith_data["Survival"]
    plt.scatter(js, jp, color="blue", s=30, alpha=0.4, edgecolors="white")
    plt.scatter(ss, sp, color="red", s=30, alpha=0.4, edgecolors="white")
    plt.xlabel("Deflection")
    plt.ylabel("Survival")
    plt.legend(labels=("Jedi", "Sith"))
else:
    plt.scatter(data['Deflection'], data['Survival'], color="green", s=30, alpha=0.4, edgecolors="white")
    plt.xlabel("Deflection")
    plt.ylabel("Survival")
    plt.legend(['Knight'])
plt.show()
```
