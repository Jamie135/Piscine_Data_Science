```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '../resources/Test_knight.csv'
data = pd.read_csv(file_path)

fig, axes = plt.subplots(nrows=6, ncols=5, figsize=(17, 15))
axes = axes.flatten()

for i, col in enumerate(data.columns):
    axes[i].hist(data[col], bins=40, color='#7fbf7f')
    axes[i].set_title(col, fontsize=14)
    axes[i].legend(['Knight'], fontsize=10)

plt.tight_layout()
plt.show()
```

```python
path = "../resources/Train_knight.csv"
df = pd.read_csv(path)

jedi = df[df['knight'] == 'Jedi']
sith = df[df['knight'] == 'Sith']

fig, axes = plt.subplots(nrows=6, ncols=5, figsize=(17, 15))
axes = axes.flatten()


for i, col in enumerate(data.columns):
    axes[i].hist(jedi[col], bins=40, alpha=0.5, color='#7f7fff')
    axes[i].hist(sith[col], bins=40, alpha=0.5, color="#ffb2b2")
    axes[i].set_title(col, fontsize=14)
    axes[i].legend(['Jedi', 'Sith'], fontsize=10)

plt.tight_layout()
plt.show()
```
