```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


df = pd.read_csv('../resources/Train_knight.csv')

df['knight'] = df['knight'].map({'Jedi': 0, 'Sith': 1})
df_scaled = StandardScaler().fit_transform(df)
df_pca = PCA().fit(df_scaled)
explained_variance = df_pca.explained_variance_
total_variance = explained_variance.sum()
variances = (explained_variance / total_variance) * 100
print(f"Variances (Percentage):\n{variances}\n")

cumulative_variances = np.cumsum(variances)
print(f"Cumulative Variances (Percentage):\n{cumulative_variances}")

plt.plot(cumulative_variances)
plt.xlabel("Number of components")
plt.ylabel("Explained variance(%)")
plt.show()
```
